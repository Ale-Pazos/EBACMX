```python
#Importar la funcion CSV para tratamiento de archivos:
import csv
#Importar OS para leer directorio actual
import os

#Obtener el directorio actual
cwd=os.getcwd()

#Imprimir directorio
print(cwd)

#Abrir archivo de lectura:
with open('Height of Male and Female by Country 2022.csv','r', encoding='utf-8') as csvfile:
    
    #Lee el archivo en el objeto csvreader
    csvreader=csv.reader(csvfile)
    
#Imprimir primeras 50 lineas
    rows=[]
    for row in csvreader:
        rows.append(row)
    print(rows[:50])
```

```python
#Imprimir todas las lineas
with open('Height of Male and Female by Country 2022.csv', 'r', encoding='utf-8') as csvfile:
    # Crear un lector CSV
    csvreader = csv.reader(csvfile)
    
    
    # Crear una lista para almacenar todas las líneas
    rows = []
    
    # Leer todas las líneas y agregarlas a la lista 'rows'
    for row in csvreader:
        rows.append(row)
    
# Imprimir todas las líneas
for row in rows:
    print(row)
```

```python
#Calcular la media, mínima y máxima altura de hombres y mujeres del data set completo

#1. importamos modulo numpy

import numpy as np

#2. Generamos en objeto height con todas las filas de la lista:
heights_global=np.array(rows)

#como referencia mostrar la primera linea del array
heights_global[0]
```

```python
#3. Muestra la columna  de "Male Height in Cm" y"Female Height in Cm"
#Recordar que cuando se usan arrays (numpy) los indices empiezan en cero!
male_heights=heights_global[0:,2]
female_heights=heights_global[0:,3]
male_heights
```

```python
# Verificación de tipo de datos

male_heights.dtype
```

```python
#4.Transformacion del array en numeros, ya que U32 significa que estoy tratando con strings

#Vamos a generar un array solamente de las alturas (valores numericos a transformar)
male_ht=male_heights[1:]
female_ht=female_heights[1:]

male_ht_float=[float(numeric_string) for numeric_string in male_ht]
#male_ht[:10] #Comprobacion
female_ht_float=[float(numeric_string) for numeric_string in female_ht]
#female_ht[:10]
```

```python


#5. Importando  modulo de Estadística:

from statistics import mean, median, mode, stdev

#6. Calculo de la media, min maxima altura de hombres y mujeres del arreglo:
mean_ht_men=mean(male_ht_float)
min_ht_men=min(male_ht_float)
max_ht_men=max(male_ht_float)

mean_ht_women=mean(female_ht_float)
min_ht_women=min(female_ht_float)
max_ht_women=max(female_ht_float)

#Reporte estadistico de los daton

print(f'En promedio a nivel mundial, la altura para hombres es {mean_ht_men:.2f} cm  y para mujeres es {mean_ht_women:.2f} cm \n') 
print(f'A nivel mundial los mínimos para la altura de hombres es {min_ht_men:.2f} cm  y para mujeres es {min_ht_women:.2f} cm \n')
print(f'A nivel mundial los máximos para la altura de hombres es {max_ht_men:.2f} cm  y para mujeres es {max_ht_women:.2f} cm \n')
```
