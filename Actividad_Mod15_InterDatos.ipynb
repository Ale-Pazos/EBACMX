## Importación de sitio web con requests e impresión del contenido (cualquier sitio)

```python
#Se debe importar libreria request para traer data de internet
import requests

#URl es el sitio
url='https://www.udemy.com/courses/business/'

data=requests.get(url)

#Se imprime la info del sitio web
if data.status_code == 200:
    # Imprime el contenido HTML de la página
    print(data.text)
else:
    # Si la solicitud no fue exitosa, muestra un mensaje de error
    print(f'Error en la solicitud. Código de estado: {data.status_code}')
```

```python
#crawling
from bs4 import BeautifulSoup
#obtiene todos los links de la pagina web
reqs=requests.get(url)
soup=BeautifulSoup(reqs.text,'html.parser')
urls=[]
for link in soup.find_all('a'):
    print(link.get('href'))
    urls.append(link.get('href')) #Esto pone las urls en una estructura tipo array
```

## Uso de librería BeautifulSoup y requests. Parseo de información

```python
#Se imprime la info del sitio web
import requests
url= 'https://www.despegar.com.mx/paquetes/'

data=requests.get(url)

#Se imprime la info del sitio web
print(data.text)
```

```python

from bs4 import BeautifulSoup
# Verifica si la solicitud fue exitosa (código de respuesta 200)

if data.status_code == 200:
    # Parsea el contenido HTML de la página web con BeautifulSoup
    soup = BeautifulSoup(data.text, 'html.parser')

    # Extrae el título de la página
    title = soup.title.string
    print(f'Título de la página: {title}')

    # Encuentra todos los enlaces (<a>) en la página y obtén sus atributos href
    links = soup.find_all('a')
    
    # Imprime los enlaces encontrados
    print('Enlaces en la página:')
    for link in links:
        href = link.get('href')
        if href:
            print(href)
else:
    print(f'Error al hacer la solicitud HTTP. Código de respuesta: {data.status_code}')
```

## Generación de un listado simple, ordenado, de los ítems que aparecen en las listas

```python
import numpy as np
url='https://www.despegar.com.mx/paquetes/'

# Realiza una solicitud HTTP para obtener el contenido de la página web
response = requests.get(url)

if response.status_code == 200:
    # Parsea el contenido HTML de la página web con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encuentra todas las listas (por ejemplo, <ul> o <ol>) en la página
    lists = soup.find_all(['ul', 'ol'])

    # Inicializa una lista vacía para almacenar los ítems de las listas
    items = []


    # Recorre todas las listas encontradas
    for lst in lists:
    # Encuentra todos los elementos de la lista (<li>)
        list_items = lst.find_all('li')

    # Agrega los ítems de la lista a la lista de ítems principal
        for item in list_items:
            items.append(item.text.strip())

    # Ordena la lista de ítems utilizando NumPy
    sorted_items = np.sort(np.array(items))

    # Imprime la lista ordenada
    for item in sorted_items:
        print(item)

else:
    print(f'Error al hacer la solicitud HTTP. Código de respuesta: {response.status_code}')
```
