 Añadir una columna “Years Playing”, que calcule el año actual menos la columna “Joined”

```python
# Año actual
current_year = 2023

# Resta 2023 a la columna 'Joined' y crea una nueva columna 'Years_Playing'
df['Years_Playing'] = current_year - df['Joined']

df
```

Generar un nuevo dataset que contenga el año (joined) y el número de jugadores (groupby)

```python
df_join=df.groupby('Joined').size().rename('Players').reset_index()

#Renombrando columna:
df_join = df_join.rename(columns={'Joined': 'Year of joining'})

df_join
```

Opcional: Generar un gráfico que contenga, por año, el número de jugadores

```python
#Grafico simple

import matplotlib.pyplot as plt
ax=df_join.plot.bar(x='Year of joining', y='Players',rot=45)
plt.show()
```
