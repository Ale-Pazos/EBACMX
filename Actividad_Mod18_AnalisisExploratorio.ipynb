```python
#Uso de librería pandas – leer archivo fifa_eda.csv 
import pandas as pd
import numpy as numpy
import os

#Cambiar el directorio actual con chdir
os.chdir(r'C:\Users\ACamacho\Desktop\EBAC\MOD18-VisualizacionDeDatos\Actividad')

#read_csv para leer el archivo .csv
df=pd.read_csv('fifa_eda.csv')

#Muestra de 3 filas
df.sample(3)
```

Elemento 1: Obtener matriz de correlación y dejarla como heat map

```python
import plotly.express as px

#Dataframe para analisis de correlacion

#Valores numéricos:
df[['Age','Overall','Potential','Value','Wage', 'International Reputation', 'Skill Moves','Height','Weight', 'Release Clause']].sample(3)
```

```python
#Generación de la matriz de correlación:

num_values=df[['Age','Overall','Potential','Value','Wage', 'International Reputation', 'Skill Moves','Height','Weight', 'Release Clause']]
corr_matrix=num_values.corr()
print(corr_matrix)
```

```python
#Heatmap de correlacion

fig=px.imshow(corr_matrix, text_auto=True)
fig.show()
```

```python
#Creando el heat map con mejoras visibles.

# Añadir etiquetas a las celdas
annotated_heatmap = corr_matrix.round(2)

# Crear el heatmap
fig = px.imshow(annotated_heatmap, 
                color_continuous_scale='Viridis',  # ELigiendo la paleta Viridis
                labels={'x':'Variables', 'y':'Variables'},  # Etiqueta los ejes
                title='Matriz de Correlación entre Variables')  # Agrega un titulo

# Valores en las celdas
fig.update_traces(text=corr_matrix.values, showscale=False)

# Personaliza el tamaño y la orientación
fig.update_layout(width=700, height=700, autosize=False)

# Agregar una barra de colores
fig.update_layout(coloraxis_showscale=True)# Ajustar la posición de la barra de colores


# Mostrar la figura
fig.show()
```
