```python
# Lista todas las columnas del DataFrame
print(df.columns)

# Asegúrate de que todas las columnas numéricas se reconozcan como números
numeric_cols = ['Year', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']

# Convierte las columnas numéricas a tipo float si es necesario
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Vuelve a generar estadísticas descriptivas
print(df.describe())
```

Totales: 1) Total de videojuegos analizados, 2) total de géneros analizados, 3) Plataformas analizadas

```python

#Total de videojuegos analizados
total_videojuegos = df.shape[0]

#Total de géneros analizados
total_generos = df['Genre'].nunique()

#Plataformas analizadas
total_plataformas = df['Platform'].nunique()

# Imprimir los resultados
print("Total de videojuegos analizados:", total_videojuegos)
print("Total de géneros analizados:", total_generos)
print("Total de plataformas analizadas:", total_plataformas)
```

```python

# Crear un gráfico de barras para explorar las plataformas
platform_counts = df['Platform'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=platform_counts.index, y=platform_counts.values, palette='viridis')
plt.title(f'Exploración de Plataformas Analizadas ({total_plataformas} plataformas)')
plt.xlabel('Plataforma')
plt.ylabel('Número de Juegos')
plt.xticks(rotation=90)
plt.show()
```

```python
# Gráfico de barras para explorar los géneros
Gender_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=Gender_counts.index, y=Gender_counts.values, palette='viridis')
plt.title(f'Exploración de Géneros ({total_generos} géneros)')
plt.xlabel('Géneros')
plt.ylabel('Número de Juegos')
plt.xticks(rotation=90)
plt.show()
```

Ventas: 1) Total de ventas por año, 2) Juegos más vendidos por año, 3) Plataformas con más ingresos por año.
Todos estos números desglosados por ingreso: NA, EU, JP, Otros y el Global, 4) Gráfico de comparación de
ventas entre géneros por año,

```python
# Limpia los valores no válidos en la columna 'Year' (pueden ser NA o inf)
df['Year'] = pd.to_numeric(df['Year'], errors='coerce', downcast='integer')
# Elimina filas con valores de 'Year' no válidos
df = df.dropna(subset=['Year'])

# Convierte la columna 'Year' en entera
df['Year'] = df['Year'].astype(int)

df.groupby(['Year'])['Global_Sales'].sum().plot.bar()
```

```python

# Encuentra el videojuego más vendido por año
most_sold_by_year = df.loc[df.groupby('Year')['Global_Sales'].idxmax()]
most_sold_by_year=most_sold_by_year[['Year','Publisher','NA_Sales', 'EU_Sales','JP_Sales', 'Other_Sales', 'Global_Sales']].reset_index(drop=True)
most_sold_by_year
```

```python
#Plataformas con más ingresos por año

most_soldPub_by_year = df.loc[df.groupby('Year')['Global_Sales'].idxmax()]
most_soldPub_by_year=most_soldPub_by_year[['Year','Publisher','NA_Sales', 'EU_Sales','JP_Sales', 'Other_Sales', 'Global_Sales']].reset_index(drop=True)
most_soldPub_by_year
```

```python
#) Gráfico de comparación de ventas entre géneros por año
# Pivotear los datos para tener géneros como columnas y sumar las ventas por año
sales_by_genre = df.pivot_table(index='Year', columns='Genre', values='Global_Sales', aggfunc='sum', fill_value=0)

# Crear un gráfico de barras apiladas
plt.figure(figsize=(20, 6))
sales_by_genre.plot(kind='bar', stacked=True)
plt.title('Comparación de Ventas entre Géneros por Año')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (en millones)')
plt.xticks(rotation=90)
plt.legend(title='Género', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Análisis de ventas a través del tiempo por región y globales

df.groupby('Year')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].mean().plot()
```

```python
import matplotlib.pyplot as plt

# Agrupar por año y calcular el promedio de ventas por región y ventas globales
sales_by_year = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].mean()

# Crear el gráfico
plt.figure(figsize=(12, 6))
plt.plot(sales_by_year.index, sales_by_year['NA_Sales'], label='NA Sales')
plt.plot(sales_by_year.index, sales_by_year['EU_Sales'], label='EU Sales')
plt.plot(sales_by_year.index, sales_by_year['JP_Sales'], label='JP Sales')
plt.plot(sales_by_year.index, sales_by_year['Other_Sales'], label='Other Sales')
plt.plot(sales_by_year.index, sales_by_year['Global_Sales'], label='Global Sales')

# Personalizar el gráfico
plt.xlabel('Año')
plt.ylabel('Ventas Promedio')
plt.title('Ventas a través del tiempo por Región y Ventas Globales')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Este gráfico puede ser útil para ver el histórico de ventas de videojuegos desde 1980 a 2020. En donde se puede observar que tuvieron un gran repunte en 1989 y en los primeros años de la década de los 90, siendo el mayor consumidor la región de Norte América. Después de 1995 se ve una caída en las ventas de estos y se mantuvieron en valores estables hasta el 2017 donde tuvieron un mínimo histórico. En casi todos los años Norte América ha sido el área donde más se ha consumido.

```python
#Gráfico de venta por género en los ultimos 5 años, para analizar cuál es el más redituable

# Filtrar el DataFrame para incluir solo los datos de los últimos 10 años
df_last_10_years = df[df['Year'] >= 2015]

df_last_10_years.groupby('Genre')['Global_Sales'].mean().sort_values(ascending=True).head(12).plot.barh()
```

Como se puede ver en el gráfico, el género de videojuegos más vendido en los ultimos años es el de shooter, siguiendo por sports y Role-Playing. Este gráfico nos sirve para ir viendo la tendencia y ver donde hay más oportunidad de mercado respecto a los géneros.
