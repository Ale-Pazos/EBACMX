1) qué género ha tenido el mayor y menor crecimiento

```python

dfsales=pd.pivot_table(df, values='Global_Sales', index=['year'],
                       columns=['Genre'],aggfunc=sum).fillna(0).reset_index()
dfsales
```

```python
#Cuadro con el porcentaje de ventas de cada género por año
import pandas as pd

# Supongamos que tienes un DataFrame llamado 'dfsales' y deseas calcular el cambio porcentual en las columnas del índice 1 al 20.
dfsalespct = dfsales.iloc[:, 1:21].pct_change()

# Reemplaza los valores "inf" y "NaN" por NaN (puedes usar cero u otro valor si lo prefieres)
dfsalespct.replace([float('inf'), float('-inf')], float('nan'), inplace=True)

# Ahora, dfsalespct no contendrá valores "inf" y "NaN".

# Verifica el DataFrame resultante
print(dfsalespct)
```

```python
dfsales.info()
```

```python
#Se obtienen los 5 generos  con mayor crecimiento promedio
dfsalespct.mean().sort_values(ascending=False).head(5)
```

```python
# Configura el gráfico de barras.
plt.figure(figsize=(12, 6))
plt.bar(dfsalespct.mean().sort_values(ascending=False).head(5).index, dfsalespct.mean().sort_values(ascending=False).head(5).values)
plt.xlabel('Género')
plt.ylabel('Cambio porcentual de ventas')
plt.title(f'Top 5 de géneros con mayor crecimiento promedio')
plt.xticks(rotation=45)

# Muestra el gráfico de barras.
plt.show()
```

```python
#Se obtienen los 5 generos  con menor crecimiento promedio
dfsalespct.mean().sort_values(ascending=True).head(5)

print("Los géneros con menor crecimiento son: ")
print(dfsalespct.mean().sort_values(ascending=True).head(5))
```

```python

dfsales_5y=dfsales.tail(3)
dfsales_5y
```

```python
#Grafico del crecimiento de promedio de cada genero

prom_5y = dfsales_5y.select_dtypes(include=['number']).mean().sort_values(ascending=False).head(10)

# Configura el gráfico de barras.
plt.figure(figsize=(12, 6))
plt.bar(prom_5y.index, prom_5y.values)
plt.xlabel('Género')
plt.ylabel('Cambio porcentual de ventas')
plt.title(f'Top10 de generos con mas ventas en los ultimos 3 años')
plt.xticks(rotation=45)

# Muestra el gráfico de barras.
plt.show()
```
