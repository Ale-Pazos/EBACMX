```python
#Importando el archivo Iris
import pandas as pd
iris=pd.read_csv("Iris.csv")
iris
```

```python
#1)

#Columnas convertidas a  float:

iris['sepal.length']=iris['sepal.length'].astype(float)
iris['sepal.width']=iris['sepal.width'].astype(float)
iris['petal.length']=iris['petal.length'].astype(float)
iris['petal.width']=iris['petal.width'].astype(float)

#Filtrando de acuerdo a las condiciones:
Cond1= iris['sepal.width'] >= 3.5
Cond2=iris['sepal.length'] > 5
df_iris=iris[Cond1 & Cond2]

#Funcion percentil50

def percentil50(columna):
    return columna.quantile(0.50)


df2_iris=df_iris.groupby('variety')["petal.width"].agg(['mean','std', percentil50])

df2_iris


#--------------------------------------------------------------------------

```

```python
#2) 
#Calcule el número de registros que se utilizaron para obtener los resultados del problema 1

num_registros=len(df_iris)

print("El número de registros utilizados fue de:", num_registros )
```

```python
#3)
#Agregue una nueva columna al DF generado en el problema 1 que sea resultado de estandarizar los valores del ancho de pétalo 
#para cada registro. Denomine dicha columna petal.width.st


#Calculando la columna petal.width.st para cada registro

df_iris

media_petalW=df_iris['petal.width'].mean()
std_petalW=df_iris['petal.width'].std()

df_iris['petal.width.st'] = (df_iris['petal.width'] - media_petalW) / std_petalW

df_iris
```

```python
#4)Mantenga en el nuevo dataframdel problema 3 las columnas "variety" y 'petal.width.st'. Renombrelas como 
#"Variedad" y "Ancho Est. Petalo"

#Renombrando
                             
df_iris = df_iris.rename(columns={"variety": "Variedad", "petal.width.st": "Ancho Est. Petalo"})

df_iris
```

```python
#5) Exportar archivo del inciso 4

df_iris.to_csv("iris2.csv")
```
