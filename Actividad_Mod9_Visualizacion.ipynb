```python
#Importando Database :
import pandas as pd
titanic=pd.read_csv('titanic.csv')
titanic
```

```python
#a) El Histograma de frecuencias para la variable edad

import matplotlib.pyplot as plt

fig, ax=plt.subplots()
ax.hist(titanic['Age'], color='skyblue') #Estableciendo la variable edad con barras de color azul claro
ax.set_xlabel("Edades")#Etiqueta del eje x
ax.set_ylabel("Conteo")#Etiqueta del eje y
plt.title('Histograma de Edades de tripulantes del Titanic') #Titulo del gráfico
 
#plt.grid(True)# rayas
plt.show()
```

```python
#b) El diagrama de caja para la variable edad
import matplotlib.pyplot as plt
fig, ax=plt.subplots()
ax.boxplot(titanic["Age"])
ax.set_xticklabels(["Edad"])
ax.set_ylabel("Conteo de personas")
plt.show()
```
